```python
import numpy as np
import matplotlib.pyplot as plt

Nombres = np.array(["Pedro", "Maria", "Jose", "Karina"])
Calificaciones = np.array([[2, 4.5, 9], [5, 3, 10], [7.5, 8.3, 9], [4, 6.5, 7]])

#np.mean() se utiliza para hacer un promedio del array 
medias_calificaciones = np.round(np.mean(Calificaciones, axis=1),2)

# Imprimir las medias de cada alumno
for i in range(len(Nombres)):   
    print(f"{Nombres[i]}: {medias_calificaciones[i]}")

# Crear una gr√°fica de barras con Matplotlib
plt.bar(Nombres, medias_calificaciones, color='skyblue')
plt.title('Medias de Calificaciones por Alumno')
plt.xlabel('Alumnos')
plt.ylabel('Media de Calificaciones')
plt.ylim(0, 10)
plt.show()
```
