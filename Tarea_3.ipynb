```python
#importamos librerias a utilizar
import numpy as np
import pandas as pd
import math as sqrt
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
 
#leer datos csv
colname=['movie_id','title','genres']
peliculas=pd.read_csv('movies_data_fix.csv',names=colname,header=None)
rating=pd.read_csv('ratings_data_fix.csv')
#imprimir datos csv
#print("peliculas base:\n",peliculas.head(20))
#print("rating base:\n",rating.head(20))
 
#eliminar año de la columna title
peliculas['year']=peliculas.title.str.extract('(\\(\\d\\d\\d\\d\\))',expand=False)
#print(peliculas['year'])
peliculas['year']=peliculas.year.str.extract('(\\d\\d\\d\\d)',expand=False)
#print("año sin parentesis:\n",peliculas['year'])
peliculas['title']=peliculas.title.str.replace(r"\s+\(.*\)","",regex=True)
#print("mostrar el title ya reemplazado:\n",peliculas['title'].head(20))
peliculas['title']=peliculas['title'].apply(lambda x:x.strip())
 
#print(peliculas.head(20))
 
#separar los valores del genero pipiline
peliculas['genres']=peliculas.genres.str.split('|')
peliculas_co=peliculas.copy()#respaldo de peliculas
for index,row in peliculas.iterrows():
    for genre in row['genres']:
        peliculas_co.at[index,genre]=1
       
peliculas_co=peliculas_co.fillna(0)
print("peliculas codificadas\n:",peliculas_co.head(20))
 
#mostrar ratingt de las peliculas
print("Rating data\n",rating.head(20))
 
#eliminamos columnas no utiles
rating=rating.drop('timestamp',axis=1)
print("Rating data nuevo\n",rating.head(20))
 
#generar datos del perfil del usuario
usuario_en=[
    {"title":"Breakfast Club","rating":2},
    {"title":"Toy Story","rating":1},
    {"title":"Jumanji","rating":5},
    {"title":"Pulp Fiction","rating":2},
    {"title":"Akira","rating":3}
]
 
#convertir usuario en dataframe
entrada_peli=pd.DataFrame(usuario_en)
print("Peliculas del usuario\n:",entrada_peli.head(20))
 
#agregamos el movie_id a las peliculas del usuario
id=peliculas[peliculas['title'].isin(entrada_peli['title'].tolist())]
entrada_peli=pd.merge(id,entrada_peli,how="inner",on=None,validate="many_to_many")
print("pelis de usuario ya con el id:\n",entrada_peli.head(20))
 
#eliminamos informacion no necesaria
entrada_peli=entrada_peli.drop('genres',axis=1).drop('year',axis=1)
print("Peliculas del user quitando col no utiles:\n",entrada_peli.head(20))
 
#codificar las peliculas del usuario para asignar metodo one hot encoding
peli_usuario=peliculas_co[peliculas_co['movie_id'].isin(entrada_peli['movie_id'].tolist())]
print("peliculas de usuario codificadas:\n",peli_usuario.head(20))
 
#otra limpieza de datos
peli_usuario=peli_usuario.reset_index(drop=True)
tabla_generos=peli_usuario.drop('movie_id',axis=1).drop('title',axis=1).drop('genres',axis=1).drop('year',axis=1)
print("Tabla de generos:\n",tabla_generos.head(20))
 
#iniciamos a modelar el algoritmo , creamos matriz de peso
#llamamos la funcion dot de pandas para multiplicar
print("tabla transpuesta de generos:\n",tabla_generos.transpose())
 
print("datos para ver el rating:\n",entrada_peli['rating'])
 
#el metodo dot multiplica cada valor de un dataframe contra los valores de
#otro dataframe y los unifica
perfil_usu=tabla_generos.transpose().dot(entrada_peli['rating'])
print("categora que el usuarios prefiere\n",perfil_usu.head(20))
 
#mostramos los generos de la tabla original
generos=peliculas_co.set_index(peliculas_co['movie_id'])
print("revisamos y validamos lista de generos\n:",generos.head(20))
 
 
#depuramos generos df
generos=generos.drop('movie_id',axis=1).drop('title',axis=1).drop('genres',axis=1).drop('year',axis=1)
print("Generos:\n",generos.head(20))
 
#Vamos a generar el proceso de recomendacion
recom=((generos * perfil_usu).sum(axis=1)) / (perfil_usu.sum())
print("Recomendaciones:\n",recom.head(20))
 
#ordernar las recomendaciones
recom=recom.sort_values(ascending=False)
print("Recomendaciones:\n",recom.head(20))
 
#tabla final de recomendacion
final=peliculas.loc[peliculas['movie_id'].isin(recom.head(20).keys())]
nfinal=final[['title']]
print("Final de recomendacion:\n",nfinal)
```
